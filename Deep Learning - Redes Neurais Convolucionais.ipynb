### 1. Redes Neurais
Uma rede neural é um sistema composto por unidades de processamento chamados neurônios interligados através de conexões sinápticas, cada uma com um valor de peso próprio e um valor de bias. Esse tipo de sistema é projetado para modelar o funcionamento do cérebro humano, visto que o conhecimento é adquirido a partir do ambiente e utiliza-se de pesos sinápticos para armazená-lo (7). Um neurônio artificial ser modelado conforme a figura 1. Existem os dados de entrada do neurônio (inputs) enviados através das sinápses, onde cada valor de entrada recebe um peso (weights). Depois os pesos e os valores de entrada são somados (transfer function), passados para uma função de ativação (activation function) que gera um valor de saída para o neurônio, caso esta função de ativação for um perceptron, a saída será binária, do contrário, se a função for uma sigmoid, unidade restrita linear (RELU) ou uma tangente hiperbólica a saída poderá ser qualquer valor real dentro de um determinado intervalo (sigmoid e tangente hiberbólica) ou de zero ao infinito para a RELU (11).<br>

### 2. Algoritmo de treinamento Gradiente Descendente

O conhecimento de uma rede neural é armazenado nos valores dos pesos sinápticos desta rede, todavia como encontrar estes valores? Para isto é utilizado um algoritmo de treinamento, o gradiente descente é um exemplo de algoritmo muito utilizado e tem como base o algoritmo backpropagation ou retropropagação (11).
Conforme Nielsen (2015) os passos deste algoritmo são:
<br>1) Receber de entrada um exemplo para treinamento como conjunto de parâmetros.
<br>2) Passar os parâmetros de entrada através da rede até a camada de saída onde resultado da rede é gerado.
<br>3) Calcular o erro comparando o resultado da rede com o valor esperado (x - y).
<br>4) Calcular o gradiente ou delta através da derivada da função de ativação em relação ao erro, para assim poder identificar a direção da função localmente.
<br>5) Retropropagação através da atualização, no sentido contrário da direção calculada no passo anterior, dos pesos sinápticos interligados a camada , pois o objetivo é diminuir o erro. Também é utilizada uma taxa para poder reduzir o tamanho da atualização dos pesos, esta taxa é conhecida como taxa de aprendizado.
<br>6) Repetir os passos 4 e 5 para todas as camadas da rede, até atingir a primeira camada. 

A figura 3 auxilia no entendimento do algoritmo gradiente descendente, onde o vale na imagem indica a função que precisa ser otimizada. Pois bem, é necessário encontrar o caminho mais fácil para descer este vale e para isso é verificado na posição atual a direção em que se possa descer. Encontrada a direção, é dado um pequeno passo (reduzido pela taxa de aprendizado) no seu sentido, até atingir a base do vale. O tamanho do passo a ser dado não pode ser tão grande ao ponto de acabar, na verdade, em um ponto mais alto que o anterior (15).

#### 2.1. Problema de Underfitting e Overfitting

Dois fenômenos podem afetar o desempenho de uma rede neural após ser treinada, o <b>underfitting</b> e <b>overfitting</b>. O algoritmo de treinamento de uma rede neural está atingindo bons resultados caso consiga diminuir o valor de custo do treinamento e tornando menor o tamanho do gap entre este custo e o custo sobre os dados de teste. Esses dois princípios são considerados os desafios centrais no aprendizado de máquina. O underfitting acontece quando o modelo não possui a capacidade suficiente para convergir ao resultado esperado, não atingindo uma taxa de erro baixa no processo de treinamento. Já o overfitting é caracterizado se o gap entre o erro gerado no conjunto de treinamento e o erro gerado no conjunto de teste é grande (6). 
<br><br>Para amenizar os efeitos causados pelo overfitting e underfitting, pode ser necessário diminuir a capacidade de operação do modelo, isto é, reduzir a capacidade de identificação de tipos diferentes de características no conjunto de treinamento. Uma rede com alta capacidade de adaptação pode acabar memorizando as características do conjunto de treinamento, obtendo resultados ruins quando aplicado em outras amostras. Por outro lado, uma rede com uma capacidade de adaptação muito baixa não conseguirá gerar o conhecimento suficiente para o conjunto de treinamento (6).


### 3. Definição do Conceito Deep Learning
Deep learning é uma nova área da inteligência artificial com o objetivo de gerar aprendizagem em vários níveis de abstração. Duas são as principais características deste conceito: 1) modelos consistentes de múltiplas camadas ou estágios de processamento de informação não linear; 2) métodos para aprendizado supervisionado e não supervisionado de representações de características em camadas mais altas e mais abstratas (4).
Deep learning é aplicado em redes neurais com várias camadas, denominadas redes neurais profundas ou deep neural networks (DNN). Uma rede neural com mais de duas camadas ocultas já pode ser considerada uma DNN (11).
<br><br>Conforme matéria publicada na revista Nature, a dificuldade no treinamento de redes profundas foi mitigado quando um eficiente algoritmo de aprendizado não supervisionado foi apresentado em dois artigos (“A fast learning algorithm for deep belief nets”, G. Hinton, S. Osindero, e Y. Teh; “Reducing the dimensionality of data with neural networks”. G. Hinton and R. Salakhutdinov) em 2006 por um grupo de pesquisadores do Canadian Institute for Advanced Research (CIFAR) (17).

#### 3 1. Qual era a dificuldade no treinamento de DNNs antes de 2006
O problema no treinamento de DNNs decorre do cálculo dos gradientes. De acordo com Nielsen (2015), o fenômeno de desaparecimento de gradientes ou vanishing gradient (VG) origina-se da retropopagação do erro das camadas finais até as camadas iniciais das DNNs e  é considerado um antigo problema no treinamento das redes neurais. Conforme o valor do gradiente, calculado na camada final da rede, é movido até as camadas iniciais o seu valor tende a ficar menor, como a multiplicação entre dois números decimais (0,1 multiplicado por 0,1 resulta em 0,01). Isto significa que os neurônios nas camadas iniciais aprendem mais lentamente que os neurônios das camadas posteriores, todavia as camadas iniciais são as mais importantes para um bom desempenho da rede neural porque elas definem as características menos abstratas. Este problema não é inevitável durante o treinamento, no entanto, sua solução tende às vezes tornar o gradiente maior nas primeiras camadas (11).

Um meio de contornar este problema surgiu, como já citado anteriormente, com o trabalho dos pesquisadores do CIFAR,
onde foram apresentados algoritmos capazes de detectar características sem precisar de dados classificados. O objetivo de fazer com que o algoritmo treine cada camada como um detector de características era poder tornar possível a reconstrução de suas atividades em camadas inferiores. Desse modo, através de um pré-treinamento de várias camadas detectoras de características, progressivamente mais complexas, os pesos de uma rede neural profunda puderam ser gerados e inicializados com dados sensíveis. Essas camadas foram compostas por uma estrutura chamada restricted boltzmann machines (RBM) empilhadas. Uma camada final de saída foi incluída na rede e o modelo pôde utilizar o algoritmo de retropropagação para o seu treinamento (17). Essa técnica funcionou perfeitamente para o reconhecimento de imagens, mesmo quando uma quantidade limitada de dados para treinamento estava disponível. Este tipo de DNN é denominada deep belief network (DBN) (17).

#### 3 2. Aprendizado em Vários Níveis de Abstração
Conforme a definição de deep learning, seu mecanismo consiste em representar o aprendizado em vários níveis de representações. Nesse sentido, Deng e Yu (2015) explicam que as camadas ocultas que estão posicionadas mais próximas à camada de entrada aprendem características de níveis mais básicos (bordas de uma imagem), identificando características locais dos dados e são mais sensíveis a mudanças nos parâmetros de entrada. As camadas que estão mais próximas à camada de saída da rede neural representam características de mais alto nível, ou seja, elas são construídas e baseadas nas camadas mais baixas e com isso são mais abstratas e constantes com relação a alterações nos parâmetros de entrada, ilustradas na figura 4 (3).

### 4.Redes Neurais Convolucionais
Redes neurais convolucionais (CNN) são um tipo de rede neural aplicadas no processamento de dados com topologia conhecida, no formato de grade. Exemplos incluem dados de séries de temporais e imagens, como uma grade 2D de pixels. O nome "rede neural convolucional" indica que a rede emprega uma operação matemática chamada convolução, um tipo especializado de operação linear. Redes convolucionais são simplesmente redes neurais que utilizam convolução em pelo menos uma das suas camadas no lugar da multiplicação de matrizes (6).
Para exemplificar o funcionamento de uma rede convolucional será utilizada a tarefa de processamento de uma imagem digital. A arquitetura de uma CNN é composta por 3 ideias: 
<br>1) <b>campos receptivos locais:</b> para cada pixel na imagem utilizada como entrada da rede é codificado a intensidade do pixel como o valor para o neurônio correspondente. Porém não serão conectados todos os pixels em todos os neurônios da camada oculta, ao invés disso, serão realizadas conexões em pequenas regiões da imagem de entrada denominados campos receptivos locais ou kernels. Cada neurônio na camada oculta pode ser considerado um holofote sobre a imagem e o conjunto de todos esses holofotes forma um filtro sobre a mesma. O filtro tem o objetivo de procurar uma característica específica em toda a imagem. A figura 6 demonstra a ideia dos holofotes sobre a imagem analisada.

2) <b>Compartilhamento de pesos e biases:</b> Como comentado anteriormente, CNNs compartilham pesos e biases. Isso porque cada neurônio na camada oculta gerado pelos campos receptivos locais possui os mesmos pesos e biases. Para que o conceito seja melhor compreendido, suponha que os pesos e os biases são tais que o neurônio na camada oculta possa selecionar uma aresta vertical em um campo receptivo local em particular. Essa habilidade também pode ser útil para outros locais na imagem e por esse motivo essas informações são compartilhadas. Uma grande vantagem no compartilhamento de pesos e biases é a redução no número de parâmetros da rede (11).

3) <b>Camadas de agrupamento:</b> além das camadas convolucionais, CNNs também possuem camadas de agrupamento ou poolings, com o objetivo de simplificar a informação na saída das camadas convolucionais. Uma camada de agrupamento pega cada saída da camada convolucional e prepara uma camada condensada de características, por exemplo, irá somar uma região de neurônios com dimensão de 2x2 da camada anterior. Um procedimento comum utilizado para a realização do agrupamento é o max-pooling, onde ele simplesmente retorna a máxima ativação da região aplicada (6).
<br>Max-pooling (figura 10) pode ser considerado como a maneira pela qual a rede verifica se determinada característica pode ser encontrada em toda uma região específica da imagem. Isso significa que não importa a localização exata desta característica na imagem, mas sim se a mesma existe ou não. Tendo como grande vantagem a redução do número de parâmetros necessários para as camadas posteriores, pois existem poucas características agrupadas (6).

### 5.TensorFlow
TensorFlow é uma biblioteca para implementação e representação de diversos algoritmos de aprendizado de máquina criada pelo Google, com lançamento para a comunidade em novembro de 2015. O projeto Google Brain iniciou em 2011 para explorar o uso de DNNs, onde nele foi desenvolvida a DistBelief, a primeira geração de um sistema distribuído para treinamento  e inferência. Posteriormente, com base na experiência adquirida nas pesquisas com a DistBelief foi criado o TensorFlow como a segunda geração do sistema. Em comparação com o DistBelief, TensorFlow é mais flexível, sua performance é melhor e ainda suporta a implementação de diversos modelos em várias plataformas de hardware (18).
<br><br>TensorFlow utiliza grafos direcionados compostos por nodos para demonstrar os cálculos realizados e possui suporte para execução de processos localmente ou distribuído em diversas máquinas. Ademais, TensorFlow  foi desenvolvido para rodar sobre CPUs ou GPUs, com a utilização da última, cálculos paralelos são executados mais rapidamente. Os sistemas operacionais suportados são MAC OS e LINUX e as linguagens de programação utilizadas no TensorFlow são Python e C++ (18).
### 6. Preparação do Ambiente de Desenvolvimento
#### 6.1 Windows:
TensorFlow tem suporte somente para Linux e Mac, então é necessário realizar a instalação de uma ferramenta para virtualização para utilizá-lo no Windows. Na página oficial do TensorFlow é recomendada a utilização da ferramenta Docker, uma plataforma de virtualização através de containers. Depois da instalação do Docker, selecione uma imagem do TensorFlow para a criação de um container. As imagens disponíveis do TensorFlow são:
- gcr.io/tensorflow/tensorflow: TensorFlow CPU binary image.
- gcr.io/tensorflow/tensorflow:latest-devel: CPU Binary image plus source code.
- gcr.io/tensorflow/tensorflow:latest-gpu: TensorFlow GPU binary image.
- gcr.io/tensorflow/tensorflow:latest-devel-gpu: GPU Binary image plus source code.

<i>docker run -it -p 8888:8888 gcr.io/tensorflow/tensorflow</i> - comando para criação de um container com a imagem do TensorFlow. ### Formato dos tensors que armazenam as imagens
Os objetos manipulados pelo TensorFlow são tensors ou arrays multidimencionais. Vamos verificar o formato de tensor utilizado para armazenar as imagens do dataset MNIST, que possui 10000 imagens de teste, 55000 imagens para treinamento e mais 5000 imagens para validação.
<br><br>Existe um tensor para cada subconjunto de imagens (teste, treinamento e validação), além de outro tensor para cada subconjunto de labels das imagens. ### Elaboração da rede neural

Após definição das variáveis de parâmetro e funções inicias, podemos partir para a modelagem da rede neural convolucional. 
TensorFlow executa todo o seu processamento pesado fora do Python para otimizar a transferência de dados, em vez de executar uma única operação independente, desse modo é necessário descrever o diagrama completo das operações.
Este processo será dividido em pequenas etapas, detalhadas uma a uma nas células seguintes:
<br>
<br>1) Definição dos placeholders que receberão os dados de entrada.
<br>2) Definição das funções que criam as variáveis para os pesos e biases.
<br>3) Definição das funções que criam as camadas convolucionais e de pooling.
<br>4) Definição das funções de custo.
<br>5) Definição dos algoritmos de treinamento.
<br>6) Criação da estrutura da rede neural convolucional.
<br>7) Implementação da função que irá treinar a rede neural.
<br>8) Definição da função para validar a rede.
<br><br>
As operações manipulam dentro do TensorFlow variáveis simbólicas que podem ser de dois tipos:
<br>
- Placeholder: tipo de variável que recebe dados de entrada para serem computados.
- Variable: é um tensor que pode ser modificado durante as operações dentro do TensorFlow.

Nesta implementação serão necessários dois placeholders, um para as imagens de entrada e outro para os labels das imagens. A primeira dimensão do tensor recebe "None", assim é possível carregá-lo com qualquer número de imagens.

### 1) Definição dos placeholders.

```python
x_pixels_imagem_entrada  = tf.placeholder(tf.float32, shape=[None, tamanho_imagem_em_pixels])
y_labels_imagem_entrada  = tf.placeholder(tf.float32, shape=[None, quantidade_labes_de_imagem])
```

### 2) Definição das funções que criam as variáveis para os pesos e biases.
As funções "variable_pesos" e "variable_biases" retornam variáveis para os pesos e biases inicializadas com valores randômicos.

```python
def variable_pesos(shape):
  inicializacao = tf.random_normal(shape, stddev=0.1)
  return tf.Variable(inicializacao)

def variable_biases(shape):
  inicializacao = tf.random_normal(shape, stddev=0.1)
  return tf.Variable(inicializacao)
```

### 3) Definição das funções que criam as camadas convolucionais e de pooling.

A função “criaCamadaConvolucional” utiliza-se de uma função do TensorFlow para criar a camada convolucional chamada “tf.nn.conv2d”, onde recebe dois parâmetros que indicam os valores de entrada da camada (x) e os valores dos pesos correspondentes (w). Nesta função pode ser especificado o tamanho do stride da convolução e o modo de padding (SAME para a saída ser no mesmo tamanho da entrada). 
<br><br>A função “criaCamadaPooling” utiliza-se da função do TensorFlow “tf.nn.max_pool” para criar a camada de pooling, onde é empregada a técnica do retorno do maior valor dentro do kernel analisado.  Esta função recebe como parâmetro os valores de entrada da camada e nela podem ser especificados o tamanho do kernel, o tamanho do stride e o modo de padding.

```python
def criaCamadaConvolucional(x, W):
  return tf.nn.conv2d(x, W, strides=[1, tamanho_stride_convolucao, tamanho_stride_convolucao, 1], padding='SAME')

def criaCamadaPooling(x):
  return tf.nn.max_pool(x, ksize=[1, tamanho_kernel_camada_pool, tamanho_kernel_camada_pool, 1],
                        strides=[1, tamanho_stride_pool, tamanho_stride_pool, 1], padding='SAME')
```

Função "camada_convolucional" define o formato da camada convolucional, recebe como parâmetros o formato do kernel (width e height), o número dos canais de entrada e a quantidade de filtros.

```python
def formato_camada_convolucional(kernel_width, kernel_height, canais, quantidade_filtros):
    return variable_pesos([kernel_width, kernel_height, canais, quantidade_filtros])
```

### 4) Definição das funções de custo.
A função de custo quantifica o desempenho do algoritmo de treinamento em relação ao seu objetivo, ou seja, se o algoritmo de treinamento está atingindo um bom desempenho se os pesos e biases que ele encontra fazem com que a função de custo tenda a zero. A função de custo quadrática, também chamada de erro quadrático médio (EQM), é um exemplo de função de custo e representa a média entre a resposta desejada da rede e resposta atual levando em consideração todos os exemplos utilizados no treinamento. 
<br>Entretanto, nessa função existe um problema quando se fazem uso de pesos e biases com valores elevados no treinamento, nesse caso, a função demora para convergir para o valor esperado na saída do neurônio [7]. Uma outra função de custo que soluciona esse problema é a cross-entropy, indicada como melhor opção e por esta razão todas as funções de custo testadas neste notebook baseiam-se nesta última [8].

```python
def cross_entropy(p_y_conv):
    f_cross_entropy = tf.reduce_mean(-tf.reduce_sum(y_labels_imagem_entrada * tf.log(p_y_conv)
                                    , reduction_indices=[1]))
    return f_cross_entropy

def cross_entropy_clip(p_y_conv):
    f_cross_entropy = tf.reduce_mean(-tf.reduce_sum(y_labels_imagem_entrada * tf.log(tf.clip_by_value(p_y_conv,1e-30,100))
                                    , reduction_indices=[1]))    
    return f_cross_entropy

def softmax_cross_entropy(p_y_conv):
    f_cross_entropy = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(p_y_conv,y_labels_imagem_entrada))
    return f_cross_entropy
```

### 5) Definição dos algoritmos de treinamento.
TensorFlow implementa diversos algoritmos de treinamento, nesta rede neural foram testados previamente 5 algoritmos, onde cada um recebe como parâmetro a taxa de aprendizado e a função de custo que precisa ser minimizada. Iremos utilizar inicialmente o algoritmo tf.train.GradientDescentOptimizer.

```python
def AdamOptimizer(custo, p_taxa_aprendizagem):
    return tf.train.AdamOptimizer(p_taxa_aprendizagem).minimize(custo)

def AdadeltaOptimizer(custo, p_taxa_aprendizagem):
    return tf.train.AdadeltaOptimizer(p_taxa_aprendizagem).minimize(custo)

def MomentumOptimizer(custo, p_taxa_aprendizagem):
    return tf.train.MomentumOptimizer(p_taxa_aprendizagem, 0.9).minimize(custo)

def RMSPropOptimizer(custo, p_taxa_aprendizagem):
    return tf.train.RMSPropOptimizer(p_taxa_aprendizagem).minimize(custo)

def SGDOptimizer(custo, p_taxa_aprendizagem):
    return tf.train.GradientDescentOptimizer(p_taxa_aprendizagem).minimize(custo)
```

### 6) Criação da estrutura da rede neural convolucional.
Para a criação das camadas da CNN é preciso definir o número de parâmetros de entrada, os pesos destinados a cada kernel e os biases (um para cada filtro de imagem).

#### Camada convolucional 1
Para a primeira camada o tensor <b>"W_convolucional_1"</b> irá armazenar os pesos da camada convolucional e terá o formato [5, 5, 1, 32], 32 filtros com kernels de dimensão 5x5 (25 pesos). O número de canais é igual a um, porque na primeira camada só existe um conjunto de valores de entrada, os pixels da imagem.
<br>O tensor <b>"b_convolucional_1"</b> deve possuir um bias para cada filtro da imagem.
<br><br>O placeholder para armazenamento dos parâmetros de entrada <b>"x_imagem_entrada"</b> terá o formato [-1,28,28,1], onde a segunda e quarta dimensões correspodem as dimensões da imagem e a quarta dimensão representa os canais de cores da imagem, como são imagens em preto e branco terá apenas um canal.

```python
W_convolucional_1 = formato_camada_convolucional(tamanho_kernel_camada_convolucional, tamanho_kernel_camada_convolucional
                                                , 1, quantidade_filtros_imagem)

b_convolucional_1 = variable_biases([quantidade_filtros_imagem])

x_imagem_entrada = tf.reshape(x_pixels_imagem_entrada, [-1,dimensao_imagem,dimensao_imagem,1])    
```

Para a finalizar a criação da camada convolucional deve-se aplicar a função de ativação aos neurônios.  Neste caso foi utilizado a sigmoid através do objeto "tf.nn.sigmoid" que recebe como parâmetro o resultado função "criaCamadaConvolucional" com a adição do tensor com os biases. 

```python
h_camada_conv_oculta_1 = tf.nn.sigmoid(criaCamadaConvolucional(x_imagem_entrada, W_convolucional_1) + b_convolucional_1)
```

#### Criação da segunda camada - pooling
Para a criação da camada de pooling, o processo é mais simples, basta passar para a função “criaCamadaPooling” a camada convolucional criada anteriormente.

```python
h_camada_pooling_1 = criaCamadaPooling(h_camada_conv_oculta_1)
```

#### Criação da terceira camada - convolucional
O processo para criação das próximas camadas, de convolução e pooling, segue os mesmos passos, porém os parâmetros de entrada da camada convolucional será agora a saída da camada de pooling <b>"h_camada_pooling_1"</b>, fazendo assim a ligação estre as mesmas.
<br>Esta camada convolucional possui o dobro de filtros, por esse motivo a variável "quantidade_filtros_imagem" foi multiplicada por 2.

```python
W_convolucional_2 = formato_camada_convolucional(tamanho_kernel_camada_convolucional, tamanho_kernel_camada_convolucional
                             , quantidade_filtros_imagem, quantidade_filtros_imagem * 2)

b_convolucional_2 = variable_biases([quantidade_filtros_imagem * 2])


h_camada_conv_oculta_2 = tf.nn.sigmoid(criaCamadaConvolucional(h_camada_pooling_1, W_convolucional_2) + b_convolucional_2)
```

#### Criação da quarta camada - pooling

```python
h_camada_pooling_2 = criaCamadaPooling(h_camada_conv_oculta_2)
```

#### Criação das camadas totalmente conectadas (FC)
Tendo em vista que o objetivo da CNN é a classificação de imagens de digitos manuscritos, será necessário ao final da rede ter uma camada capaz de gerar o resultado da predição, para esta tarefa é empregada uma camada softmax. Desse modo, após a última camada de pooling é adicionado uma camada totalmente conectada (FC) para fazer o processamento de toda a imagem e por último, como camada de saída, a softmax. O primeiro passo é selecionar a dimensão do tensor de saída da última camada da rede (neste caso h_camada_pooling_2), essa informação será útil para gerar as próximas camadas FC. A dimensão encontrada foi armazenada na variável <b>pool_shape</b>.

#### Criação da primeira camada FC
Com o valor de "pool_shape" foram criadas as variáveis para os pesos e bisases dos neurônicos da primeira camada FC. A variável para os pesos possui o formato da dimensão da saída da camada anterior, por exemplo, se a última camada for um pooling com saída de <b>dimensão 7x7 através de 64 filtros</b>, a matriz de pesos terá o formato <b>[7x7x64 , numero_neuronios_camada_fc]</b>. A variável para os biases é criada com o mesmo formato do número total de neurônios da camada FC.

```python
pool_shape = h_camada_pooling_2.get_shape()    

W_camada_fc1 = variable_pesos([int(pool_shape[1] * pool_shape[1] * pool_shape[3]), numero_neuronios_camada_fc]) 
b_camada_fc1 = variable_biases([numero_neuronios_camada_fc])

h_camada_pooling_2_flat = tf.reshape(h_camada_pooling_2, [-1, int(pool_shape[1] * pool_shape[1] * pool_shape[3])])
h_camada_fc1 = tf.nn.sigmoid(tf.matmul(h_camada_pooling_2_flat, W_camada_fc1) + b_camada_fc1)

taxa_dropout = tf.placeholder(tf.float32)

h_camada_fc1_dropout = tf.nn.dropout(h_camada_fc1, taxa_dropout)
```

<br>Para realizar a ligação entre a camada FC e a última camada de pooling, é preciso redimensionar o tensor de retorno do pooling, dessa forma o tensor redimensionado torna-se um array sequenciado com o total de valores <b>7x7x64=3136</b> (64 filtros de dimensão 7x7). Tendo as variáveis para os pesos e biases especificadas pode-se criar a camada FC, no entanto, há um detalhe a mais na criação desse tipo de camada, antes de aplicar a função sigmoid aos neurônios, faz-se necessário efetuar a multiplicação entre as matrizes dos pesos e valores de entrada, a função <b>tf.nn.conv2d</b> usada para as camadas convolucionais já faz este trabalho, todavia, para as camadas FC este cálculo deve ser feito manualmente. 
<br><br>A API torna mais fácil a tarefa de multiplicação entre os tensors que contêm os pesos e valores de entrada, através da utilização da função “tf.matmul”. Aplicada a multiplicação, basta passar o resultado para a função de ativação e somar com os biases. Feito isso a primeira camada totalmente conectada está quase pronta. 
<br><br>O último detalhe é por conta da aplicação da técnica de dropout na camada através da função "tf.nn.dropout" que recebe como parâmetro a saída da camada e a taxa de dropout. Está técnica é utilizada para reduzir o overfitting da rede e será explicada com mais detalhes no final do notebbok.


#### Criação da segunda camada totalmente conectada
Recebe como entrada a saída da camada FC anterior.

```python
W_camada_fc2 = variable_pesos([numero_neuronios_camada_fc, numero_neuronios_camada_fc]) 
b_camada_fc2 = variable_biases([numero_neuronios_camada_fc])

h_camada_fc2 = tf.nn.sigmoid(tf.matmul(h_camada_fc1_dropout, W_camada_fc2) + b_camada_fc2)

h_camada_fc2_dropout = tf.nn.dropout(h_camada_fc2, taxa_dropout)
```

#### Criação da camada de saída
O processo de criação da camada de saída da rede é semelhante ao processo da camada FC anterior, a variável para a matriz de pesos é dada pelo número de neurônios da camada anterior pela quantidade de labels da imagem, este também define o número de biases. Porém ao invés de utilizar a função sigmoid é utilizado a função softmax, gerando as probabilidades proporcionais aos labels conforme a predição correta.

```python
W_softmax_fc_3 = variable_pesos([numero_neuronios_camada_fc, quantidade_labes_de_imagem]) 
b_softmax_fc3 = variable_biases([quantidade_labes_de_imagem])

y_camada_saida_final = tf.nn.softmax(tf.matmul(h_camada_fc2_dropout, W_softmax_fc_3) + b_softmax_fc3)
```

O objeto <b>tf.nn.softmax</b> produz o resultado da aplicação da função softmax a um tensor de entrada, normalizando o resultado em probabilidades, para que a soma de todas saídas seja igual a 1. O tensor de saída de uma softmax terá o mesmo formato do tensor de entrada. No exemplo abaixo é aplicada a softmax a um tensor com quatro elementos.

```python
a = tf.constant(np.array([[9.0, 5.0, 9.0, 9.0]]))
print (sessao.run(tf.nn.softmax(a)))
```

### Seleção da função de custo, otimizador e cálculo da acurácia
Para finalizar a criação do modelo da CNN devemos definir a função de custo, o otimizador e o cálculo responsável pela acurácia do modelo.
<br>Nesta primeira simulação de exemplo foi definida a função de custo cross_entropy passando como parâmetro a saída da rede neural. O otimizador escolhido foi o "GradientDescentOptimizer", onde passa-se a função de custo que deve ser minimizada e a taxa de aprendizagem. 
<br><br>Para avaliar a acurácia da rede utiliza-se a função “tf.equal”, que compara a igualdade entre dois valores, neste caso o índice com valor de probabilidade mais alto em uma dimensão do tensor. O índice com maior probabilidade é selecionado com a função “tf.argmax”, aplicada no tensor que contêm os labels das imagens e no tensor com as probabilidades de saída da rede. “tf.equal” gera como retorno uma lista de valores booleanos.
<br><br>Com a lista de valores booleanos gerada, aplica-se um cast para gerar uma lista de binários. Isso é necessário para o emprego da função “tf.reduce_mean” que computa a média entre uma lista de valores e assim geramos a porcentagem da acurácia.

```python
funcao_custo = cross_entropy(y_camada_saida_final)
otimizador = SGDOptimizer(funcao_custo, taxa_aprendizagem)
predicoes_finais = tf.equal(tf.argmax(y_camada_saida_final,1), tf.argmax(y_labels_imagem_entrada,1))
calcula_acuracia = tf.reduce_mean(tf.cast(predicoes_finais, tf.float32))
```

### 7) Implementação da função que irá treinar a rede neural
<br>
O treinamento da CNN é realizado pela função "treinarRedeNeural()", abaixo são explicados os principais comandos  executados dentro dela:
- <b>sessao.run(tf.initialize_all_variables())</b> :  são inicializadas todas as variáveis existentes no TensorFlow.
- <b>batch = mnist.train.next_batch(tamanho_mini_batch)</b> :  gera um mini-batch com imagens selecionadas randomicamente para o treinamento.  
- <b>accuracia.eval(feed_dict={....</b> :  a cada 10 ciclos de treinamento a acurácia da rede neural é avaliada. A função "eval" executa e retorna, no formato de um nparray, o valor de objetos, neste caso o objeto "calcula_acuracia".
- <b>feed_dict={x_pixels_imagem_entrada:batch[0], y_labels_imagem_entrada: batch[1]....</b> : o parâmetro feed_dict permite a alteração de valores de tensors dentro do diagrama do TensorFlow durante a chamada das funções "eval" e "run". Para o cálculo da acurácia são alterados os valores dos placeholders "x_pixels_imagem_entrada", que recebe os pixels das imagens, e "y_labels_imagem_entrada", que recebe os labels correspondentes.
- <b>_, l, predictions = sessao.run([otimizador, funcao_custo, predicoes_finais]....</b> :a função "run" executa todo o diagrama computacional do TensorFlow e retorna os resultados das operações indicadas no argumento "fetches", alimentado por uma lista de operações, que neste exemplo são três [otimizador, funcao_custo, predicoes_finais]. 

```python
def treinarRedeNeural():
    
    global acuracias_durante_treinamento
    global custo_durante_treinamento
    global x_range_acuracia
    global x_range_custo
    global valor_taxa_dropout
    global tamanho_mini_batch
    
    #inicializa as variáveis
    acuracias_durante_treinamento = []    
    custo_durante_treinamento = []
    x_range_acuracia = []
    x_range_custo = []
    tempo0 = dt.datetime.now()
    
    print ("Tempo de inicio do treinamento " + str(tempo0))
    
    #inicializa todas as variáveis dentro do TensorFlow
    sessao.run(tf.initialize_all_variables())
    l = 0.0        
    
    #inicia o treinamento da rede neural
    for i in range(ciclos_treinamento+1):
        
        #gera um mini-batch com imagens selecionadas randomicamente
        batch = mnist.train.next_batch(tamanho_mini_batch)        
        if i%10 == 0:      
            
            #calcula a acurácia da rede neural
            acuracia_gerada = calcula_acuracia.eval(feed_dict={
                x_pixels_imagem_entrada:batch[0], y_labels_imagem_entrada: batch[1], taxa_dropout: 1.0})                            
            
            #armazena a porcentagem da acurária
            acuracias_durante_treinamento.append(acuracia_gerada)
            x_range_acuracia.append(i)  
            
            #armazena o custo calculado
            x_range_custo.append(i)        
            custo_durante_treinamento.append([np.round(l,5)])            
            print("step %d, acuracia no treinamento %g"%(i, acuracia_gerada*100)+"%")      
         
        #executa o treinamento da rede neural
        feed_dict = {x_pixels_imagem_entrada: batch[0], y_labels_imagem_entrada: batch[1], taxa_dropout: valor_taxa_dropout}                                  
        _, l, predictions = sessao.run([otimizador, funcao_custo, predicoes_finais],feed_dict=feed_dict)            

    print("Finalizou o treinamento")
    tempo1 = dt.datetime.now()
    print ("Tempo de treinamento " + str(tempo1 - tempo0))
```

### 8) Definição da função para validar a rede
Função "validaCNN" é utilizada para validação e geração da acurácia da rede através das imagens de teste.

### Simulação com a função de ativação Relu
<br>
Vamos alterar o digrama da rede neural no TensorFlow para utilizar a função RELU. Como a função de ativação impacta toda a rede neural precisamos recriar todas as camadas, substituindo a função "tf.nn.sigmoid" pela função "tf.nn.relu".

```python
W_convolucional_1 = formato_camada_convolucional(tamanho_kernel_camada_convolucional, tamanho_kernel_camada_convolucional
                                               , 1, quantidade_filtros_imagem)
b_convolucional_1 = variable_biases([quantidade_filtros_imagem])
x_imagem_entrada = tf.reshape(x_pixels_imagem_entrada, [-1,dimensao_imagem,dimensao_imagem,1])    
h_camada_conv_oculta_1 = tf.nn.relu(criaCamadaConvolucional(x_imagem_entrada, W_convolucional_1) + b_convolucional_1)
h_camada_pooling_1 = criaCamadaPooling(h_camada_conv_oculta_1)
W_convolucional_2 = formato_camada_convolucional(tamanho_kernel_camada_convolucional, tamanho_kernel_camada_convolucional
                             , quantidade_filtros_imagem, quantidade_filtros_imagem * 2)
b_convolucional_2 = variable_biases([quantidade_filtros_imagem * 2])
h_camada_conv_oculta_2 = tf.nn.relu(criaCamadaConvolucional(h_camada_pooling_1, W_convolucional_2) + b_convolucional_2)
h_camada_pooling_2 = criaCamadaPooling(h_camada_conv_oculta_2)
pool_shape = h_camada_pooling_2.get_shape()    

W_camada_fc1 = variable_pesos([int(pool_shape[1] * pool_shape[1] * pool_shape[3]), numero_neuronios_camada_fc])    # 7 * 7 * 64, 1024 #(7x7x1024=50176)    
b_camada_fc1 = variable_biases([numero_neuronios_camada_fc])

h_camada_pooling_2_flat = tf.reshape(h_camada_pooling_2, [-1, int(pool_shape[1] * pool_shape[1] * pool_shape[3])])
h_camada_fc1 = tf.nn.relu(tf.matmul(h_camada_pooling_2_flat, W_camada_fc1) + b_camada_fc1)

taxa_dropout = tf.placeholder(tf.float32)

h_camada_fc1_dropout = tf.nn.dropout(h_camada_fc1, taxa_dropout)

W_camada_fc2 = variable_pesos([numero_neuronios_camada_fc, numero_neuronios_camada_fc]) 
b_camada_fc2 = variable_biases([numero_neuronios_camada_fc])

h_camada_fc2 = tf.nn.relu(tf.matmul(h_camada_fc1_dropout, W_camada_fc2) + b_camada_fc2)

h_camada_fc2_dropout = tf.nn.dropout(h_camada_fc2, taxa_dropout)
    
W_softmax_fc_3 = variable_pesos([numero_neuronios_camada_fc, quantidade_labes_de_imagem]) 
b_softmax_fc3 = variable_biases([quantidade_labes_de_imagem])

y_camada_saida_final = tf.nn.softmax(tf.matmul(h_camada_fc2_dropout, W_softmax_fc_3) + b_softmax_fc3)

funcao_custo = cross_entropy(y_camada_saida_final)
otimizador = SGDOptimizer(funcao_custo, taxa_aprendizagem)
predicoes_finais = tf.equal(tf.argmax(y_camada_saida_final,1), tf.argmax(y_labels_imagem_entrada,1))
calcula_acuracia = tf.reduce_mean(tf.cast(predicoes_finais, tf.float32))
```

### Redução de overfitting com dropout
<br>
Dropout é uma técnica poderosa de regularização (redução do overfitting) de baixo custo computacional. Sua estratégia não se baseia na alteração da função de custo. Ao invés disso, altera diretamente a estrutura da rede neural propriamente utilizando-se de uma prática de bagging. 
<br><br>Bagging consiste no treinamento e avaliação de várias redes para o mesmo dado de exemplo. Portanto, dropout treina o conjunto de todos os modelos diferentes de subredes que podem ser formados removendo unidades das camadas ocultas da rede aleatoriamente. Esse processo é repetido inúmeras vezes, restaurando os neurônios removidos e escolhendo um novo conjunto de neurônios para serem deletados, estimando o gradiente e atualizando os valores dos pesos sinápticos e biases da rede. Desse modo, a rede poderá aprender pesos e biases sobre diferentes modelos (6).

### Regularização L2
<br>
A estratégia de regularização L2 foca na alteração da função de custo, acrescentando um termo de regulariação. Este termo consiste  na soma dos quadrados de todos os pesos da rede, escalado por um parâmetro de regularização “λ”, em que λ > 0 (6).
<br>O principal objetivo da regularização L2 é fazer com que a rede prefira gerar valores pequenos para os pesos, sendo que pesos como valores mais altos só serão permitidos caso melhorarem substancialmente a função de custo. Sendo assim, a regularização L2 pode ser vista como uma forma de compromisso entre gerar pesos com valores baixos e minimizar a função de custo original (LISA LAB, 2015). A importância relativa aos dois elementos de compromisso depende do valor estipulado para o termo de regularização: quando o valor deste termo é pequeno é preferível minimizar a função de custo original, porém quando  o seu valor é mais elevado, concentra-se na geração de valores baixos para os pesos sinápticos (11). 
<br><br>A questão que permeia a geração de valores baixos para os pesos é a possibilidade de reduzir a complexidade do modelo, fornecendo uma explicação mais simples e poderosa para os dados. Com valores baixos para os pesos o comportamento da rede não vai ser muito alterado caso sejam realizadas pequenas alterações aleatórias nos dados de entrada, portanto reprimindo a rede de aprender ruídos nos dados (LISA LAB, 2015).
<br>Em resumo uma rede regularizada não visa aprender características específicas dos dados, ao invés disso, procura generalizar o modelo aprendendo características que são vistas frequentemente nos dados. Ao contrário de uma rede neural não regularizada com pesos sinápticos elevados, onde gera um modelo complexo baseado no ruído apresentado pelos  dados (11). 
<br><br>Para aplicar a regularização L2 iremos seguir três passos:
<br>1) Calcular soma dos quadrados de todos os pesos da rede através da função "tf.nn.l2_loss".
<br>2) Multiplicar o resultado do cálculo anterior pelo fator de regularização (beta) dividido pelo número das imagens de treinamento sobre o tamanho do mini_batch.
<br>3) Adicionar à função de custo o termo de regularização gerado no passo 2.

```python
regularizadores = (tf.nn.l2_loss(W_camada_fc1) + tf.nn.l2_loss(W_camada_fc2) + tf.nn.l2_loss(W_softmax_fc_3))

funcao_custo += regularizadores * (beta_regularizacao / (total_imagens_treinamento / tamanho_mini_batch))

otimizador = AdamOptimizer(funcao_custo, taxa_aprendizagem)

ciclos_treinamento      = 1000
treinarRedeNeural()

tamanho_batch_validacao = 2000
validaCNN(tamanho_batch_validacao)
```
